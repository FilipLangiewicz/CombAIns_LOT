```python
train['label'] = pd.Categorical(
    list(zip(
        train['graphic_design'],
        train['top_1_section'],
        train['top_2_section'],
        train['top_3_section']
    ))
).codes
```

```python
train_clicked_1 = train[train['clicked'] == 1]
```

```python
print(f"Train shape: {train.shape}, Train clicked 1 shape: {train_clicked_1.shape}")
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train = train.drop(columns=['clicked', 'label', 'graphic_design', 'top_1_section', 'top_2_section', 'top_3_section'])
y_train = train[['label', 'clicked']]
```

```python
train_clicked_1.head(1)
```

```python
x_train_clicked_1 = train_clicked_1.drop(columns=['clicked','graphic_design', 'top_1_section', 'top_2_section', 'top_3_section', 'label'])
y_train_clicked_1 = train_clicked_1[['label', 'clicked']]
```

```python
_, x_valid_clicked_1, _, y_valid_clicked_1 = train_test_split(x_train_clicked_1,
                                                              y_train_clicked_1, 
                                                              test_size=0.6, 
                                                              random_state=42, 
                                                              stratify=y_train_clicked_1['label']
                                                              )
```

```python
x_train = x_train.drop(x_valid_clicked_1.index)
y_train = y_train.drop(y_valid_clicked_1.index)
```

```python
x_train.to_csv("data/x_train.csv", index=False)
y_train.to_csv("data/y_train.csv", index=False)

x_valid = x_valid_clicked_1.to_csv("data/x_valid.csv", index=False)
y_valid = y_valid_clicked_1.to_csv("data/y_valid.csv", index=False)
```
