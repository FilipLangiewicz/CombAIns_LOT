```python
import pandas as pd
data = pd.read_csv('data/hackaton_students_train.csv')
data
```

```python
data['label'] = data['graphic_design'].astype('str') + \
                 data['top_1_section'].astype('str') + \
                 data['top_2_section'].astype('str') + \
                 data['top_3_section'].astype('str')
data
```

```python
data.drop(columns=['graphic_design', 'top_1_section', 'top_2_section', 'top_3_section'], inplace=True)
data['label'] = data['label'].astype('category').cat.codes
data
```

```python
data['index'] = data.index
data
data.to_csv('data/F/data_tmp.csv', index=False)
```

```python
new_rows = []

# Iterujemy przez kad obserwacj w oryginalnej ramce danych
for index, row in data.iterrows():
    original_clicked = row['clicked']
    original_layout = row['label']
    
    # Dodajemy kolumn 'original_index' z numerem wiersza
    
    # Dla kadej wartoci layout od 0 do 9, z wyjtkiem tej, ktra jest ju w wierszu
    for layout_value in range(10):
        if layout_value != original_layout:
            # Tworzymy nowy wiersz z kopiami wszystkich kolumn
            new_row = row.copy()
            new_row['label'] = layout_value
            # Zmieniamy warto 'clicked' na 0, niezalenie od oryginalnej wartoci
            new_row['clicked'] = 0
            new_rows.append(new_row)

# Dodajemy nowo wygenerowane wiersze do oryginalnej ramki danych
new_rows_df = pd.DataFrame(new_rows)
expanded_data = pd.concat([data, new_rows_df], ignore_index=True)

# Wydrukowanie nowej ramki danych
print(expanded_data)
```

```python
from sklearn.model_selection import train_test_split

# Wycigamy unikalne indeksy
unique_indices = expanded_data['index'].unique()
print(unique_indices)
# Dzielimy unikalne indeksy na train/valid
train_indices, valid_indices = train_test_split(
    unique_indices,
    test_size=0.2,
    random_state=23
)

# Tworzymy finalne zbiory danych na podstawie podziau indeksw
X_train = expanded_data[expanded_data['index'].isin(train_indices)].drop(columns=['clicked'])
X_valid = expanded_data[expanded_data['index'].isin(valid_indices)].drop(columns=['clicked'])
y_train = expanded_data[expanded_data['index'].isin(train_indices)]['clicked']
y_valid = expanded_data[expanded_data['index'].isin(valid_indices)]['clicked']

# Zapis do plikw
X_train.to_csv('data/F/X_train.csv', index=False)
X_valid.to_csv('data/F/X_valid.csv', index=False)
y_train.to_csv('data/F/y_train.csv', index=False)
y_valid.to_csv('data/F/y_valid.csv', index=False)
```

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

# Krok 1: Automatyczne wykrywanie kolumn kategorycznych i numerycznych
categorical_columns = data.select_dtypes(include=['object', 'category']).columns
numerical_columns = data.select_dtypes(include=['int64', 'float64']).columns

# Krok 2: Usuwanie skorelowanych cech tylko dla kolumn numerycznych
def remove_highly_correlated_features(data, numerical_columns, threshold=0.7):
    # Korelacja tylko dla numerycznych kolumn
    corr_matrix = data[numerical_columns].corr().abs()
    upper_triangle = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    
    # Znajd kolumny, ktre s bardzo skorelowane
    to_drop = [column for column in upper_triangle.columns if any(upper_triangle[column] > threshold)]
    
    # Usu te kolumny z danych
    data = data.drop(columns=to_drop)
    return data, to_drop

# Przetwarzanie danych: usuwanie skorelowanych cech
data_cleaned, dropped_columns = remove_highly_correlated_features(data, numerical_columns, threshold=0.7)

# Zaktualizowane listy kolumn po usuniciu skorelowanych cech
numerical_columns = [col for col in numerical_columns if col not in dropped_columns]

# Krok 3: Tworzymy pipeline do przetwarzania danych
#  - Kolumny numeryczne: standaryzacja
#  - Kolumny kategoryczne: OneHotEncoding

# Pipeline dla kolumn numerycznych (standaryzacja)
numeric_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),  # Imputacja brakujcych wartoci (jeli wystpuj)
    ('scaler', StandardScaler())  # Skalowanie
])

# Pipeline dla kolumn kategorycznych (OneHotEncoding)
categorical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),  # Imputacja brakujcych wartoci
    ('onehot', OneHotEncoder(drop='first', sparse_output=False))  # One-Hot Encoding
])

# Krok 4: Tworzymy ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_pipeline, numerical_columns),  # Przetwarzanie kolumn numerycznych
        ('cat', categorical_pipeline, categorical_columns)  # Przetwarzanie kolumn kategorycznych
    ])

# Zastosowanie transformacji dla danych treningowych
data_transformed = preprocessor.fit_transform(data_cleaned)

# Uzyskujemy nazwy kolumn po One-Hot Encoding
categorical_columns_encoded = preprocessor.transformers_[1][1].get_feature_names_out(categorical_columns)

# Tworzymy ostateczny DataFrame
final_columns = np.concatenate([numerical_columns, categorical_columns_encoded])
data_final = pd.DataFrame(data_transformed, columns=final_columns)

# Wywietlenie przetworzonych danych treningowych
print(data_final)

# Teraz przetwarzamy dane walidacyjne X_valid
# Zakadajc, e X_valid jest w tej samej strukturze co data_cleaned, wykonujemy tylko transformacj:
X_valid_transformed = preprocessor.transform(data_valid)

# Uzyskujemy nazwy kolumn po One-Hot Encoding dla danych walidacyjnych
categorical_columns_encoded_valid = preprocessor.transformers_[1][1].get_feature_names_out(categorical_columns)

# Tworzymy ostateczny DataFrame dla X_valid
final_columns_valid = np.concatenate([numerical_columns, categorical_columns_encoded_valid])
X_valid_final = pd.DataFrame(X_valid_transformed, columns=final_columns_valid)

# Wywietlenie przetworzonych danych walidacyjnych
print(X_valid_final)
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
from sklearn.metrics import accuracy_score
from L_score import L_score

# Funkcja do trenowania modelu i walidacji po kadej epoce
def train_and_evaluate_model(model, X_train, X_valid, y_train, y_valid, epochs=10):
    for epoch in range(epochs):
        # Trenowanie modelu
        model.fit(X_train, y_train)
        
        # Przewidywanie na zbiorze walidacyjnym
        y_pred_probs = model.predict_proba(X_valid)
        
        # Dodajemy indeks oraz kolumn 'label' z X_valid
        X_valid_with_probs = data_v.copy()
        X_valid_with_probs['pred_clicked_prob'] = y_pred_probs[:, 1]  # Prawdopodobiestwo klasy 1

        grouped = X_valid_with_probs.groupby('index', group_keys=False)

        # Wybr wiersza z najwikszym prawdopodobiestwem w kadej grupie
        selected_rows = grouped.apply(lambda group: group.loc[group['pred_clicked_prob'].idxmax()]).reset_index(drop=True)

        # Tworzymy wektor y_pred z wartociami 'label' wybranych wierszy
        y_pred = selected_rows['label'].values
        
        tmp = data_v[['index']].copy()
        tmp['y_valid'] = y_valid.values

        # Grupowanie po 'index' i przypisanie maksymalnej wartoci z y_valid jako label
        tmp = tmp.groupby('index', as_index=False)['y_valid'].max()
        tmp.rename(columns={'y_valid': 'clicked'}, inplace=True)
        
        data_tmp = pd.read_csv('data/F/data_tmp.csv')
        data_tmp['index'] = data_tmp.index
        clicked = data_tmp[data_tmp['index'].isin(valid_indices)]['label']
        tmp['label'] = clicked.values
        
        # Ocena modelu przy uyciu funkcji L_score
        score = L_score(tmp, y_pred)
        
        # Wywietlanie wyniku dla kadej epoki
        print(f"Epoch {epoch+1}/{epochs}, L_score: {score}")

y_train = pd.read_csv('data/F/y_train.csv')
y_train = y_train['clicked']  # Upewnij si, e y_train jest seri
y_valid = pd.read_csv('data/F/y_valid.csv')
y_valid = y_valid['clicked']  # Upewnij si, e y_valid jest seri

# # Model 1: Logistic Regression
# print("Trening modelu Logistic Regression")
# log_reg_model = LogisticRegression(max_iter=100)  # W przypadku wikszej liczby iteracji, aby unikn bdw
# train_and_evaluate_model(log_reg_model, X_train, X_valid, y_train, y_valid, epochs=10)

# Model 2: Random Forest Classifier
print("\nTrening modelu Random Forest")
rf_model = RandomForestClassifier(n_estimators=100, random_state=42, criterion='gini')
train_and_evaluate_model(rf_model, X_train, X_valid, y_train, y_valid, epochs=5)

# # Model 3: XGBoost
# print("\nTrening modelu XGBoost")
# xgb_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
# train_and_evaluate_model(xgb_model, X_train, X_valid, y_train, y_valid, epochs=10)
```
